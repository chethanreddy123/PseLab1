```python
import pandas as pd
data = pd.read_csv("State_Region_corrected.csv")
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
data.plot(x = "Area (km2)", y = "National Share (%)" )
plt.show()
```

```python
data.plot(kind="barh", y="National Share (%)", x="State / Union territory (UT)",figsize=(10,10))
plt.show()
```

```python
data.plot(kind="pie", y="National Share (%)", x="State / Union territory (UT)",figsize=(10,10))
plt.show()
```

```python
# Now Let's Try to predict the Power generation in percentage 
# When the Area is given....
```

```python
plt.scatter(data['Area (km2)'], data['National Share (%)'])
plt.xlabel("Area (km2)")
plt.ylabel("National Share (%)")
plt.show()
```

```python
# As the Graph is almost a kind of straight line
# Linear Regression is the best model, to train the data set
```

```python
from sklearn.linear_model import LinearRegression
Model = LinearRegression()
```

```python
# Here I have used all data set to train the model 
# as the data set is very less..
Model.fit(data[['Area (km2)']], data[['National Share (%)']])
```

```python
res = Model.predict([[342239]])
print("The predict value of powerGeneration: ",round(res[0][0],2))
print("The actually Value from the dataset is: ",10.55)
```

```python
# Let's see the best fit line that Our Linear Regression Model Made
print("The Linear Equation is: ")
print("y = {} * x ".format(Model.coef_[0][0]))
```

```python
y_predict = []
for i in data['Area (km2)']:
    y_predict.append(Model.predict([[i]]))
Final = []
for i in range(len(y_predict)):
    Final.append(y_predict[i][0][0])
print("The Power Generation Values % by Model: ")
print()
print(Final)
```

```python
plt.plot(data[["Area (km2)"]],Final)
plt.scatter(data[["Area (km2)"]],data[['National Share (%)']],color = "red")
plt.xlabel("Area (km2)")
plt.ylabel("National Share (%)")
plt.title("# Our Model has predicted the best fit line for the given data set:- ")
plt.show()
```
